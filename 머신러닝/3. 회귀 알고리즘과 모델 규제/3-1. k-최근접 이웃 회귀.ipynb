```python
import matplotlib.pyplot as plt

plt.scatter(perch_length, perch_weight)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(perch_length, perch_weight, random_state=42)
# 길이에 따른 무게를 예측할 거임
```

#### sklearn에 데이터를 넣어주려면 ( ?, ?)로 2차원 배열이 되어야 해서 억지로 만들어 줄 거임 

```python
test_array = np.array([1,2,3,4])
print(test_array.shape)
print(test_array)
```

```python
test_array = test_array.reshape(2,2)# 원소 개수 맞춰야함 (2 x 2 = 4)
print(test_array.shape)
print(test_array)  # 2차원 배열 형태로
```

```python
print(X_train.shape, X_test.shape)
X_train = X_train.reshape(-1, 1) # 첫번쨰 크기를 -1로 하면 나머지 원소 개수로 모두 채우란 의미(첫번쨰 크기 유지)
X_test = X_test.reshape(-1, 1)
print(X_train.shape, X_test.shape)
```

### 결정 계수 (R^2) == score

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor()
knr.fit(X_train, y_train)
```

```python
print(knr.score(X_test, y_test))
```

#### R^2 = (타깃-예측)^2 의 합 / (타깃-타깃평균)^2 의 합   
R^2 이 얼마나 좋은지 이해하기 어려우므로 다른 값으로도 계산해보자

```python
from sklearn.metrics import mean_absolute_error
# 테스트 세트에 대학 예측
test_pred = knr.predict(X_test)

# 평균값 절대 오차
mae = mean_absolute_error(y_test, test_pred)
print(mae) # 타깃값과 10g 정도의 오차가 있음을 알 수 있다
```

### 과대적합 vs 과소적합
- 과대적합 : 훈련 세트에서 점수는 매우 높지만 테스트 세트에서 점수가 매우 나쁨(훈련 세트에만 적합)
- 과소적합 : 훈련 세트보다 테스트 세트의 점수가 높거나 둘 다 너무 낮은 경우(모델이 단순해 적절한 훈련 x)

```python
# 테스트 세트가 아닌 훈련세트의 R^2 점수 확인
print(knr.score(X_train, y_train)) # 테스트 세트의 점수가 더 높음 -> 과소적합
```

#### 해결하기 위해 모델을 좀 더 복잡하게 만들면 된다. -> 훈련 세트에 더 잘 맞게 만들기
최근접 이웃의 수 k 를 줄이면 된다. -> 이웃 개수 줄이면 훈련세트에 있는 국지적인 패턴에 민감해지고, 이웃의 개수를 늘리면 데이터 전반에 있는 일반적인 패턴을 따를 것이다.

```python
# 이웃 3개로 설정
knr.n_neighbors = 3
# 모델 다시 훈련
knr.fit(X_train, y_train)
print(knr.score(X_train, y_train))
print(knr.score(X_test, y_test))
# 과소적합 문제 해결
```

### 확인 문제

```python
knr = KNeighborsRegressor()
# np에서 arange는 앞 뒤 둘 다 포함( 5~45까지 x좌표)
x = np.arange(5, 45).reshape(-1, 1)
# n= 1, 5, 10일 때 예측 결과를 그래프로 그린다
for n in [1, 5, 10]:
    knr.n_neighbors = n
    knr.fit(X_train, y_train)
    
    pred = knr.predict(x)
    
    plt.scatter(X_train, y_train)
    plt.plot(x, pred)
    plt.show() # 이웃의 수를 늘릴수록 훈련세트와의 적합성이 떨어지는 것을 볼 수 있음
```
